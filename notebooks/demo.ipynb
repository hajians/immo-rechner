```python
fig, ax = plt.subplots()
[
    ProfitCalculator.from_raw_data(
        yearly_income=100_000,
        monthly_rent=1500,
        facility_monthly_cost=350.0,
        owner_share=0.5,
        repayment_amount=repayment_amount,
        yearly_interest_rate=0.035,
        initial_debt=450_000,
        depreciation_rate=0.02,
        purchase_price=500_000,
    )
    .simulate(n_years=10, to_pandas=True)
    .plot(
        x="year", y="cashflow", style="*-", ax=ax, label=f"repayment {repayment_amount}"
    )
    for repayment_amount in range(500, 1700, 200)
]
```
